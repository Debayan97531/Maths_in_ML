```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

*Plotting a vector*

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-7,7)
plt.ylim(-7,7)
plt.show()
```

```python
plt.quiver(0,0,5,6, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,5,6, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,-6, scale_units='xy' ,angles='xy',scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Adding two vectors**

```python
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,-3,-4])

sum=vector_1+vector_2
print(sum)
```

**Subtraction of two vectors**

```python
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,-3,-4])

diff=vector_1-vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,-4, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,5,7, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Multiply with a scalar**

```python
vector_1=np.asarray([0,0,3,-5])


vec_2=2*vector_1

vec_3=-0.5*vector_1

print(vec_2)
print(vec_3)
```

```python
plt.quiver(0,0,3,-5, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,6,-10, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1.5,2.5, scale_units='xy', angles='xy', scale=1, color='g')

plt.xlim(-11,11)
plt.ylim(-11,11)
plt.show()
```
